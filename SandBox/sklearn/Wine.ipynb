```python
from sklearn.datasets import load_wine
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = load_wine()
```

```python
print(data.DESCR)
```

```python
data.keys()
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
target = data.target
random = 11
```

```python
df.head()
```

```python
target
```

```python
df_corr = df.corr()
sns.heatmap(df_corr, annot = True, )
plt.siz
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn.model_selection import train_test_split as tts
dtc = DTC()
from sklearn import tree
```

```python
x1, x2, y1, y2 = tts(df, target, test_size=0.25, random_state=random)
```

```python
dtc.fit(x1, y1)
```

```python
dtc.score(x2, y2)
```

```python
tree.plot_tree(dtc)
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC

param = {
    'n_estimater':[10*a for a in range(1,11)],
    'depth':[b for b in range(1,11)]
}
scores = []
for a in param.get('n_estimater'):
    for b in param.get('depth'):
        rfc = RFC(random_state=random,n_estimators=a,max_depth =b )
        rfc.fit(x1,y1)
        scores.append((rfc.score(x2,y2),a))
```
