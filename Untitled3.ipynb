```python
import pandas as pd
import numpy as np

german = pd.read_csv("http://freakonometrics.free.fr/german_credit.csv")
```

```python
from sklearn.linear_model import LinearRegression as LR
from sklearn.model_selection import train_test_split as tts
train = german.iloc[:,1:]
target = german.iloc[:,0]
lr = LR()
x_train, x_test, y_train, y_test = tts(train, target, test_size = 0.2 , shuffle=True)
```

```python
lr.fit(x_train, y_train)
lr.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

for a in range(1,100):
    knc = KNeighborsClassifier(n_neighbors=a)
    knc.fit(x_train, y_train)
    if(knc.score(x_test, y_test)) > 0.72:
        print(a)
```

```python
knc = KNeighborsClassifier(n_neighbors=56)
knc.fit(x_train, y_train)
knc.score(x_test, y_test)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=16)
```

```python
cross_val_score(lr, x_train, y_train, scoring='accuracy', n_jobs=1, cv = k_fold)
```
