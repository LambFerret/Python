```
import numpy as np
import pandas as pd
wine = pd.read_csv('https://bit.ly/wine-date')

data = wine[['alcohol','sugar','pH']].to_numpy()
target = wine[['class']].to_numpy()

from sklearn.model_selection import train_test_split as tts
train_input,test_input,train_target,test_target = \
tts(data, target, test_size=0.2, random_state=42)

```

```
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier as rfc
rf = rfc(n_jobs=-1, random_state=42)
scores = cross_validate(rf, train_input, train_target,
                        return_train_score = True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```
rf = rfc(oob_score=True, n_jobs=-1, random_state=42)
rf.fit(train_input, train_target)
print(rf.oob_score_)
```

```
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier as hgbc
hgb = hgbc()
scores = cross_validate(hgb, train_input, train_target,
                        return_train_score=True)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```
from sklearn.inspection import permutation_importance
hgb.fit(train_input, train_target)
result = permutation_importance(hgb, train_input, train_target,
                                n_repeats = 10, random_state=42, n_jobs=-1)
print(result.importances_mean)

result2 = permutation_importance(hgb, test_input, test_target,
                                n_repeats = 10, random_state=42, n_jobs=-1)
print(result2.importances_mean)
```

```
from xgboost import XGBClassifier as xgbc
xgb = xgbc(tree_method='hist', random_state=42)
scores = cross_validate(xgb, train_input, train_target,
                        return_train_score=True)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```
