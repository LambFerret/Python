```
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split as tts

cancer = load_breast_cancer()
x = cancer.data
y = cancer.target
x_train_all, x_test, y_train_all, y_test = tts(x,y,stratify = y, test_size=0.2)
x_train, x_val, y_train, y_val = tts(
    x_train_all,y_train_all, stratify = y_train_all, test_size=0.2)
# train+val = train_all, train_all + test = x
```

```
from sklearn.preprocessing import PolynomialFeatures as pf
poly = pf()
poly.fit(x_train)
train_poly = poly.transform(x_train)
print(train_poly.shape)
```

```
from sklearn.linear_model import LinearRegression as lrr
lr = lrr()
lr.fit(train_poly,y_train)
print(lr.score(train_poly,y_train))
```
