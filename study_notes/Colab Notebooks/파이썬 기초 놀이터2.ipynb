```
def print_rxn(a):
  b = int(len(a)/5)
  for x in range(b + 1):
    print(a[x*5:x*5+5])
```

```
def printmxn(string,num):
  many = int(len(string)/num)
  for x in range(many +1):
    print(string[x*num:x*num+num])
```

```
printmxn("가나다라육마바사아자차카타치키쵸",6)
```
